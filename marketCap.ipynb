```python
import yfinance as yf
from pandas_datareader import data as pdr
import csv
import json
yf.pdr_override()
```

```python
# importar los symbols
f = open('symbols.json')
symbols = json.load(f)
```

```python
symbols["symbols"]
```

```python
result = []
```

```python
for sym in symbols["symbols"]:
  print(f"Getting -> {sym}")
  marketCap = pdr.get_quote_yahoo(sym)["marketCap"]
  print(f'Add new element {sym} -> {marketCap[0]}')
  item = [sym, marketCap[0]]
  result.append(item)
  print('Element added!')
```
